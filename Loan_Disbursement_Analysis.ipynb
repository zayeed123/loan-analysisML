```python
#Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df1['Loan_Status']=le.fit_transform(df1['Loan_Status'])
```

```python
df1 = pd.get_dummies(data=df1, columns=['Married'], drop_first=False)
```

```python
df1.head(10)
```

```python
#Logistic Regression

x=df1[['ApplicantIncome','CoapplicantIncome','LoanAmount','Credit_History']]
y=df1.Loan_Status
```

```python
#Model

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(xTrain, yTrain)
y_pred = logreg.predict(xTest)
```

```python
cnf_matrix = metrics.confusion_matrix(yTest, y_pred)
cnf_matrix
```

```python
print('Accuracy:',metrics.accuracy_score(yTest, y_pred))
print('Precision:',metrics.precision_score(yTest, y_pred))
print('Recall:',metrics.recall_score(yTest, y_pred))
```
